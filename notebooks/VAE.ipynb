```python
data = pd.read_csv("../data/qst_dataset.csv")
N = data.shape[0]
data.head()
# clip any small values to zero
eps = 1e-10

#create boolean mask
mask = data.abs() < eps
data[mask] =0

X_cols = []
y_cols = []

N_x = 36
N_y = 32
for i in range(N_x):
    X_cols.append(f"x{i}")
for i in range(N_y):
    y_cols.append(f"y{i}")

X = data[X_cols]
y = data[y_cols]

#split the data
X_train, X_test, y_train, y_test=  train_test_split(X, y, train_size=0.7, random_state=42)
```
